```python
def createOutgoingLinks(data):
    result = pd.DataFrame(columns=["Origin", "Destination"])
    tempMap = {}
    #taking care of the pages which send links
    for index, row in data.iterrows():
        origin = row["FromNodeId"]
        destination = row["ToNodeId"]
        if origin not in tempMap:
            tempMap[origin] = [destination]
        else:
            tempMap[origin].append(destination)
    #Now it is time to take care of pages who dont have outgoing links
    for index, row in data.iterrows():
        destination = row["ToNodeId"]
        if destination not in tempMap:
            tempMap[destination] = []
    result = pd.DataFrame(tempMap.items(), columns = ["Origin", "Destination"])
    return result
```

```python
#Now lets put everything in a function so that we can perform an iteration
def pageRank(data, numberOfIterations):
    #creating a links rdd containing information about the pages, pages it
    #points to and initial rank = 1
    links = createOutgoingLinks(data)
    links["Rank"] = 1
    
    #this process will be repeated n (numberOfIterations) number of times
    for i in np.arange(0, numberOfIterations):
        #it is time to calculate the number of outgoing links, and the rank
        #which is distributed accordingly
        tempLength = []
        rankDistributed = []
        for row in links.itertuples():
            length = len(row[2])
            if length  == 0:
                rank = 1
            else: 
                rank = row[3] / length
            tempLength.append(length)
            rankDistributed.append(rank)
        links["NumberOfOutGoingLinks"] = tempLength
        links["RankDistributed"] = rankDistributed

        #Now that we know what each page will get, it is time to add the ranks
        #so first we need to create a new dataframe, preferably a map which will keep all the ranks
        #it will be getting

        newMap = {}
        for row in links.itertuples():
            destination = row[2]
            for place in destination:
                if place not in newMap:
                    newMap[place] = row[5] * 0.85
                else:
                    newMap[place] += row[5] * 0.85
        for keys in newMap:
            newMap[keys] += 0.15

        #now that we have the updated PageRanks of each page
        #it is time to update the new ranks in the dataframe and the ranks distributed
        newRank = []
        ranksDistributed = []
        for row in links.itertuples():
            page = row[1]
            rank = -1
            if page in newMap:
                rank = newMap[page]
            else:
                #no page links to this page, so 0.15
                rank = 0.15
            #now to take care of the ranks distributed
            if row[4] == 0:
                ranksDistributed.append(rank)
            else:
                ranksDistributed.append(rank/row[4])
            newRank.append(rank)
        links["Rank"] = newRank
        links["RankDistributed"] = ranksDistributed
    return links
```
